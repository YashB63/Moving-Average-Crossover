```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = yf.download('^GDAXI', start='2022-1-1')#start='2023-1-1' at this only one buying and selling is obtained. Let's change the year to 2022
```

```python
df
```

```python
df['Moving_Average_20'] = df['Adj Close'].rolling(20).mean()
df['Moving_Average_50'] = df['Adj Close'].rolling(50).mean()
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
df = df[['Adj Close','Moving_Average_20','Moving_Average_50']]
```

```python
df
```

```python
# we are just checking fot the buying case:

# If the moving average of the 20 days is above the moving average of 50 days but isn't the day before =>  this is the buying signal
# If the moving average of the 50 days is above the moving average of the 20 days but isn't the day before => this is the selling signal

buy = []
sell = []

for i in range(len(df)):
    if df.Moving_Average_20.iloc[i] > df.Moving_Average_50.iloc[i] \
    and df.Moving_Average_20.iloc[i-1] < df.Moving_Average_50.iloc[i-1]:
        buy.append(i)
    elif df.Moving_Average_20.iloc[i] < df.Moving_Average_50.iloc[i] \
    and df.Moving_Average_20.iloc[i-1] > df.Moving_Average_50.iloc[i-1]:
        sell.append(i)
```

```python
buy # row numbber where these conditions are fulfilled
```

```python
sell
```

```python
plt.figure(figsize=(14,6))
plt.plot(df['Adj Close'], label='Asset price', c='orange', alpha=0.5)
plt.plot(df['Moving_Average_20'], label='Moving_Average_20', c='k', alpha=0.9)
plt.plot(df['Moving_Average_50'], label='Moving_Average_50', c='magenta', alpha=0.9)
plt.scatter(df.iloc[buy].index,df.iloc[buy]['Adj Close'],marker='^', color='g', s=130)
plt.scatter(df.iloc[sell].index,df.iloc[sell]['Adj Close'],marker='v', color='r', s=130)
plt.legend()
plt.show()
```
